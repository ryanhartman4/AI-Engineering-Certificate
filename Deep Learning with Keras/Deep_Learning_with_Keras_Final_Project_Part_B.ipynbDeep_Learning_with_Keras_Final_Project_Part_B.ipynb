```python
# Importing libraries
import tensorflow
from tensorflow import keras
import pandas as pd
from google.colab import files
from keras import Sequential
from keras.layers import Dense 
from sklearn.model_selection import train_test_split
```

```python
# Uploading files 
files.upload()
```

```python
# Creating dataframe
path = '/content/concrete_data.csv'
ConcreteDF = pd.read_csv(path)
ConcreteDF.head()
```

```python
# Creating predictor and target dataframes
Predictors = ConcreteDF.drop('Strength',axis=1).copy()
Target = ConcreteDF[['Strength']].copy()
```

```python
from sklearn import preprocessing
# Normalizing predictor columns
Scaler = preprocessing.MinMaxScaler()
names = Predictors.columns
scaled = Scaler.fit_transform(Predictors)
Scaled_Predictors = pd.DataFrame(scaled, columns=names)
```

```python
# Creating train/test split
X_train, X_test, Y_train, Y_test = train_test_split(Scaled_Predictors, Target, test_size=.30,random_state=420)
```

```python
# Creating variables
num_cols = X_train.shape[1]

# Initializing Regression Model with 10 layers plus output layer
model = Sequential()
model.add(Dense(10,activation='relu',input_shape=(num_cols,)))

# Adding output layer
model.add(Dense(1))
```

```python
# Checking model creation
model.summary()
```

```python
# Compiling model with Adam optimizer and MSE loss function 
model.compile(optimizer='Adam',loss='mean_squared_error', metrics=['mean_squared_error'])
```

```python
# Fitting model to training data (using 50 epochs)
model.fit(X_train,Y_train, epochs=50)
```

```python
# Evaluating on Test Data
results = model.evaluate(X_test,Y_test)
print('')
print('Current Mean Squared Error', results[1])
```

```python
# Creating list for all 50 MSE's
MSEs = []
```

```python
# Looping over train test split, training, and evaluation for 50 times. Adding MSEs to list

# Initializing counter 
counter = 0
while counter < 50:
    # Creating train/test split
    X_train, X_test, Y_train, Y_test = train_test_split(Scaled_Predictors, Target, test_size=.30,random_state=counter)
    # Fitting model
    model.fit(X_train,Y_train, epochs=50, verbose=0)
    # Evaluating model
    results = model.evaluate(X_test, Y_test)
    print('Current Trial =', counter+1)
    print('Current Trial Mean Squared Error', results[1])
    print('___________________')
    # Adding MSE to list
    MSEs.append(results[1])
    # Incrementing counter
    counter += 1
```

```python
# Calculating average MSE and Standard Deviation of MSEs
mean = sum(MSEs)/len(MSEs)
variance = sum([((x - mean) ** 2) for x in MSEs]) / len(MSEs)
STDdev = variance ** 0.5
print('Average MSE:',mean)
print('Standard Deviation of MSEs:', STDdev)
```

Looks like the average MSE is much lower than the average MSE from the data before normalization.
